**Onehotencoding with many categories in a feature**

```python
import pandas as pd
import numpy as np
```

```python

df=pd.read_csv('/content/drive/MyDrive/datasets/mercedesbenz.csv',usecols=["X0","X1","X2","X3","X4","X5","X6"])
```

```python
df.head()
```

```python
for columns in df.columns:
  print(columns,':',len(df[columns].unique()))
```

```python
# lets examin how many columns will be obtain after one hot encoding these variables i.e. 163-7 and if it's many that will lead to curse of dimentionality
pd.get_dummies(df,drop_first=True).shape
```

https://proceedings.mlr.press/v7/niculescu09/niculescu09.pdf


we will use 10 most frequent labels(category) of a variable and use one hot encoding on that

```python
#lets find the top 10 most frequent categories for vaiable x1
lst_10=df.X1.value_counts().sort_values(ascending=False).head(10).index
lst_10=list(lst_10)
```

```python
lst_10
```

```python
for categories in lst_10:
    df[categories]=np.where(df['X1']==categories,1,0)
```

```python
df.head()
```

```python
# get whole set of dummy variables, for all the categorical variable

def one_hot_encoding_top_X(df,variable,top_x_variable):
  # function to create dummy variable for most frequent variables
  for labels in top_x_variable:
    df[variable + '_' + labels] = np.where(df[variable]==labels,1,0)
    
```

```python
lst_10_x2=df.X2.value_counts().sort_values(ascending=False).head(10).index
lst_10_x2=list(lst_10)
```

```python
one_hot_encoding_top_X(df,'X2',lst_10_x2)
```
